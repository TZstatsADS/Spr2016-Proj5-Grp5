```python
nick = {}
with open('name_list.txt', 'rb') as f:
    for line in f:
        if line.strip().partition("\t")[-1] != '':
            nick[line.strip().partition("\t")[0]] = [line.strip().partition("\t")[0]]+line.strip().partition("\t")[-1][1:-1].split(', ')
        else:
            nick[line.strip().partition("\t")[0]] = [line.strip().partition("\t")[0]]
f.close()
```

```python
import pickle
with open("names_with_nick_ep1", "wb") as f:
    pickle.dump(nick, f)
```
