```python
# Extract names from text and save the frequency of names as a dictionary
def extract_entities(text):
    mylist = []
    for sent in nltk.sent_tokenize(text):
        for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent))):
            if hasattr(chunk, 'label'):
                if chunk.label() == 'PERSON': 
                    #print chunk.label, ' '.join(c[0] for c in chunk.leaves())
                    mylist += [c[0] for c in chunk.leaves()]
    mydict = {}
    for x in mylist:
        if x in mydict.keys():
            mydict[x] += 1
        else:
            mydict[x] = 1
    return mydict
```

```python
# Read books and save as text
def read_book(book):
    with open (book,"r") as f:
        line=f.readlines()
    text='\n'.join(line)
    text=text.decode("utf8")
    return text
```

```python
text1 = read_book("book_1.txt")
dict1 = extract_entities(text1)
```
